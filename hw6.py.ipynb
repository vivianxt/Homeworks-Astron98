# Van Der Waals Gas

```python
P, V = np.meshgrid(np.linspace(1, 10), np.linspace(10,30)) #units is bar
#V = np.linspace(10,30) #units in liters
R = 0.083144
a = 16.02
b = 0.1124
n = 1
T = ((P + (a*(n**2))/(V**2))*(V - (n*b)))/(n*R)


plt.figure()
plt.imshow(T, extent = [1,10, 10, 30] ,cmap = 'inferno', origin = 'lower', vmin = 0, vmax = 4000) #extent is axis #origin think about pillars #vmin and vmax defines colorbar
plt.xlabel('Pressure')
plt.ylabel('Volume')
plt.colorbar()
plt.show()
```

# Monte Carlo

```python
#increasing it by one is a good way to count 
pow = [3,4,5,6]
xin = []  #empty list
yin = []
xout = []
yout = []
for p in pow:
    num_in = 0
    for i in range(10**p):
        x = np.random.rand(1) #puts in another bracket which is why [0] comes in 
        y = np.random.rand(1)
        if x**2 + y**2 <= 1:    #think of circle radius of 1
            xin.append(x)
            yin.append(y)
            num_in = num_in + 1
        else:
            xout.append(x)
            yout.append(y)
print(4*num_in/10**p)

        
```

```python
plt.figure()#A is pi
plt.xlabel('x')
plt.ylabel('y')
plt.scatter(xin,yin)
plt.scatter(xout,yout)
plt.show()
```
