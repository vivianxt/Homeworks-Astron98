```python
col = tableglt_df['State']
#col == ['Wyoming', 'Nebraska', 'South Dakota']
newtglt_df = tableglt_df[(col == 'Wyoming') | (col=='Nebraska')| (col=='South Dakota')]
newtglt_df
```

```python
glt1b_df = newtglt_df.groupby(['dt'])[["AverageTemperature"]].mean().reset_index()
glt1b_df[['dt', 'AverageTemperature']]
```

```python
x = glt1b_df['dt']
y = glt1b_df['AverageTemperature']

plt.figure(figsize = (20,5))
plt.title('Global Land Temperature by Date')
#sns.scatterplot(data = glt1b_df, x = 'dt', y = 'AverageTemperature')
plt.plot(x, y)
plt.xlabel('Date')
plt.ylabel('Average Temperature')
plt.xticks(rotation = 90)
plt.show()
```
